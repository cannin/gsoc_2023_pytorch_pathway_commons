```python
df = pd.read_csv("/content/drive/MyDrive/data_mrna_seq_v2_rsem.txt", sep="\t")
```

```python
patients = df.copy()
patients.head()
```

```python
patients = patients.drop('Entrez_Gene_Id', axis=1)
```

```python
patients.isnull().sum()
```

```python
patients.shape
```

```python
patients = patients.dropna()
```

```python
patients.head()
```

```python
identifiers = patients['Hugo_Symbol'].unique()
print(len(identifiers))
```

```python
import random

unique_symbols = list(patients['Hugo_Symbol'].unique())
sampled_symbols = random.sample(unique_symbols, 20)

for symbol in sampled_symbols:
    print(symbol)
```

```python
# Check for duplicate 'Hugo_Symbol' values
duplicates = patients['Hugo_Symbol'].duplicated(keep='last')

# Drop rows with duplicate 'Hugo_Symbol' or same symbol as the previous row
new_df = patients[~duplicates]

# Verify the updated DataFrame shape
print(new_df.shape)
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
new_df_transposed = new_df.set_index('Hugo_Symbol').transpose()
new_df_transposed.head()
```

```python
new_df_transposed.shape
```

```python
# Get the list of index names in df2_transposed
index_names = new_df_transposed.columns.tolist()

# Rename the columns in df1 to match the index names in df2_transposed
df1 = used_df.rename(columns=dict(zip(new_df.columns[1:], index_names)))
```

```python
df1.head()
```

```python
df1.shape
```

```python
# Merge the datasets based on the patient identifiers
merged_df = pd.merge(df1, new_df_transposed, left_on='Sample Identifier', right_index=True)
```
