```python
import pandas as pd
```

```python
data_sample = pd.read_csv("/content/drive/MyDrive/data_clinical_sample.txt", sep="\t")
```

```python
data_sample.head()
```

```python
ds1 = data_sample.drop([0,1,2,3])
ds1.head()
```

```python
ds1.shape
```

```python
ds1["#Patient Identifier"]
```

```python
# Extract the columns of interest
sample1 = ds1['Cancer Type Detailed']
sample2 = ds1['Cancer Type']

# Check if the two columns contain the same values
are_equal = sample1.equals(sample2)

if are_equal:
    print("The columns contain the same values.")
else:
    print("The columns do not contain the same values.")
```

```python
ds1.isnull().sum()
```

```python
cleaned_ds = ds1.dropna(axis=1, how='all')
cleaned_ds.isnull().sum()
```

```python
cleaned_ds = cleaned_ds.fillna(method='ffill')
cleaned_ds.isnull().sum()
```

```python
cleaned_ds = cleaned_ds.drop(columns='Cancer Type')
cleaned_ds.sample(10)
```

```python
cleaned_ds.shape
```

```python
data_patient = pd.read_csv("/content/drive/MyDrive/data_clinical_patient.txt", sep="\t")
data_patient.head()
```

```python
dcp = data_patient.drop([0,1,2,3])
dcp.head()
```

```python
dcp.isnull().sum()
```

```python
# Extract the columns of interest
dcp1 = dcp['Subtype']
dcp2 = dcp['TCGA PanCanAtlas Cancer Type Acronym']

# Check if the two columns contain the same values
are_equal = dcp1.equals(dcp2)

if are_equal:
    print("The columns contain the same values.")
else:
    print("The columns do not contain the same values.")
```

```python
# Check for mismatches between the columns
mismatched_values = dcp1[~dcp1.isin(dcp2)]

# Print the mismatched values
print("Mismatched values:")
print(mismatched_values)
```

```python
# Extract the columns of interest
dcp1 = dcp['Subtype']
dcp2 = dcp['TCGA PanCanAtlas Cancer Type Acronym']

# Check if the two columns contain the same values
are_equal = dcp1.equals(dcp2)

if are_equal:
    print("The columns contain the same values.")
else:
    print("The columns do not contain the same values.")
```
