# Process Pathway Commons Data

```python
path = "PathwayCommons12.reactome.hgnc.sif.gz"


#specify compression type because the file is compressed
df = pd.read_csv(path, sep="\t", compression="gzip", header=None, names=["Source", "InteractionType", "Target"])
```

```python
# Copy of the original Pathway Commons data before filtering
df_org = df
```

```python
df = df.drop(columns='InteractionType')
```

```python
df.head()
```

```python
all_nodes = sorted(list(set(df['Source']) | set(df['Target'])))
len(all_nodes)
```

```python
# Get number of common nodes between Pathway Commons and newly merged cBioPortal data
used_nodes = sorted(list(set(X.columns) & set(all_nodes)))
len(used_nodes)
```

```python
with open('cbioportal_nodes.txt', 'w') as f:
    f.write('\n'.join(used_nodes))
```

```python
# Filter rows in the DataFrame where the 'Source' column values are in 'used_nodes'
source_values = df['Source'].values
mask = np.isin(source_values, used_nodes)
df = df.loc[mask]

# Filter rows in the DataFrame where the 'Target' column values are in 'used_nodes'
source_values = df['Target'].values
mask = np.isin(source_values, used_nodes)
df = df.loc[mask]

# Reset the index of the DataFrame after filtering rows
df = df.reset_index(drop=True)

# Display the first two rows of the DataFrame 'X'
X.head(2)
```

```python
len(set(df['Source']) | set(df['Target']))
```

```python
# Get the unique values from the 'Source' and 'Target' columns and combine them into a set
columns_to_select = sorted(set(df['Source']).union(set(df['Target'])))

# Select the columns in 'X' that are present in 'columns_to_select'
X = X[columns_to_select]

# Display the first two rows of the updated 'X' DataFrame
X.head(2)
```

# Creating PyTorch Geometric Edge Index from the Pathway Commons Data

```python
!pip install torch-geometric
```

```python
import torch
```

```python
source_nodes = df['Source'].tolist()
target_nodes = df['Target'].tolist()


#Create a dictionary to map each unique node to a unique index
node_to_index = {node: index for index, node in enumerate(X.columns)}

# Map the source and target nodes to their corresponding indices
source_indices = [node_to_index[node] for node in source_nodes]
target_indices = [node_to_index[node] for node in target_nodes]
```

```python
# Convert the source and target indices to a PyTorch tensor
edge_index = torch.tensor([source_indices, target_indices], dtype=torch.long)
```

```python
X.shape, y.shape
```

```python
# Set the fixed indices for splitting using 60:20:20 ratio
train_index = int(0.6 * X.shape[0])
val_index = int(0.6 * X.shape[0]) + int(0.2 * X.shape[0])

# Split the data into train, validation, and test sets
X_train = X[:train_index]
X_val = X[train_index:val_index]
X_test = X[val_index:]
y_train = y[:train_index]
y_val = y[train_index:val_index]
y_test = y[val_index:]
```
